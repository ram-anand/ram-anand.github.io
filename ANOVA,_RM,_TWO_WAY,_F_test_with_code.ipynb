## Fishers F-Test
 
#### F-distribution

>The F-distribution is formed by the ratio of two independent chi-square variables divided by their respective degrees of freedom.
- Select a random sample of sample variance $s_{1}^2$ of size $n_{1}$ from a normal population, having a standard deviation equal to $\sigma_{1}$, $\chi_{1}^2$ is the chi-square statistic for the sample drawn from population 1.
- Select an independent random sample of sample variance $s_{2}^2$ of size $n_{2}$ from a normal population, having a standard deviation equal to $\sigma_{2}$,  $\chi_{2}^2$ is the chi-square statistic for the sample drawn from population 2 . 
- If both sample comes from same population then $\mu_{1}= \mu_{2}$
- Mean of F distribution = $\mu = \frac{df(num)}{df(denom)}\, -1$

>\begin{align}F = \frac{s_{1}^2/\mu_{1}^2}{s_{2}^2/\mu_{2}^2} = \frac{\chi_{1}^2/df_{1}}{\chi_{2}^2/df_{2}}\end{align}

>\begin{align}\frac{df_{1}s_{1}^2}{\sigma^2} \sim \chi_{1}^2, \frac{df_{2}s_{2}^2}{\sigma^2} \sim \chi_{2}^2\end{align}

>\begin{align}df_{1} = (n_{1}-1), df_{2} = (n_{2}-1)\end{align}

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/F-distribution_pdf.svg/330px-F-distribution_pdf.svg.png" alt="Normal"></img>

**Characterisitics**
- The F-values are all non-negative
- The distribution is non-symmetric
- The mean is approximately 1
- There are two independent degrees of freedom, one for the numerator, and one for the denominator.
- Distribution is positively skewed towards right. When numerator dof and denominator dof becomes larger, the distribution becomes bell shaped and symmetric.

#### F-Test for Variances

- Null hypothesis: No difference in population variances
- Alternative hypothesis: There is a difference

The F-test is designed to test if two population variances are equal. It does this by comparing the ratio of two variances. So, if the variances are equal, the ratio of the variances will be 1.

#### F-statistic 

 - Is a ratio of two variances (Variance is nothing but measure of dispersion, it tells how far the data is dispersed from the mean). F statistic accounts corresponding degrees of freedom to estimate the population variance. The larger variance should always be placed in the numerator.
 - F-statistic is almost similar to t statistic. t-test states a single variable is statistically significant or not whereas F test states a group of variables are statistically significant or not.

> \begin{align} F = \frac{\text{Expained Variability}}{\text{Unexplained Variability}} = \frac{\text{Between Group Variability}}{\text{Within Group Variability}} = \frac{MS_{between}} {MS_{within}} = \frac{MS_{regression}} {MS_{error}} = \frac{\sigma_{1}^2} {\sigma_{2}^2} = \frac{s_{1}^2} {s_{2}^2}\end{align}

The F-test is typically used to answer one of the following questions:
1. Do two samples come from populations with equal variances?
2. Does a new treatment or process reduce the variability of some current treatment or process?
3. Are the the means between two populations significantly different or not?

### F-test for linear regression model 
  - To tests **any** of the independent variables (predictor) in a multiple linear regression are significant or not. 
  - To compare two competing regression models (model1 with p1 predictor, model2 with p2 predictor) in their ability to **explain** the variance in the dependent/target variable
  - It also indicates a linear relationship between dependent variable and at least one of the independent variable.

>\begin{align}F = \frac{(SSE_{model1}-SSE_{model2})/(p_{2}-p_{1})}{SSE_{model2}/(n-p_{2})} \end{align}



### Assumptions

1. Normality – Samples in each group should follow normal distribution, only needed for small sample sizes, say n < 20 per group.
2. Equal Variances – The variances of the populations that the samples come from should be equal. Homogeneity is only needed if sample sizes are very unequal (Levene's Test).
3. Independence – The observations in each group are independent of each other and the observations within groups were obtained by a random sample.
4. Same Sample Size: The treatment groups should have the same sample size.


### Types of F-test

1. **Pouplation variance is known:**
F test is to find out whether the two independent estimates of population variance differ significantly.

- Null Hypothesis(H0): $\sigma_{1}^2 = \sigma_{2}^2$ (the population variances are equal)
- Alternate Hypothesis(H1): $\sigma_{1}^2 \neq \sigma_{2}^2$ (the population variances are not equal)

 In this case F ratio is
\begin{align} F = \frac{\sigma_{1}^2}{\sigma_{2}^2}, σ_{1}^2 > σ_{2}^2\end{align}
where
\begin{align} \sigma_{1}^2 = \frac{\sum (X_{1}-\mu_{1})^2}{n_{1}}, \sigma_{2}^2 = \frac{\sum (X_{2}-\mu_{2})^2}{n_{2}}\end{align}
  
larger estimate of variance always be in numerator and smaller estimate of variance in denominator 
 - DF of larger variance (i.e numerator) $a=n_{1}-1$
 - DF of smaller variance (i.e denominator) $b=n_{2}-1$

2. **Pouplation variance is unknown:**
To find out whether the two samples drawn from the normal population having the same variance. 

- Null Hypothesis(H0): $s_{1}^2 = s_{2}^2$ (the unbaised sample variances are equal)
- Alternate Hypothesis(H1): $s_{1}^2 \neq s_{2}^2$ (the unbaised sample variances are not equal)

In this case F ratio is

\begin{align}F = \frac{s_{1}^2}{s_{2}^2}, s_{1}^2 > s_{2}^2\end{align} where 
\begin{align}s_{1}^2 = \frac{\sum (X_{1}-\bar{x_{1}})^2}{n_{1}-1}, s_{2}^2 = \frac{\sum (X_{2}-\bar{x_{2}})^2}{n_{2}-1}\end{align}

larger estimate of variance always be in numerator and smaller estimate of variance in denominator 
 - DF of larger variance (i.e numerator) $a=n_{1}-1$
 - DF of smaller variance (i.e denominator) $b=n_{2}-1$


### Steps to conduct F-test
1. Choose the test: Note down the independent variables and dependent variable and also assume the samples are normally distributed
2. Calculate the F statistic, choose the highest variance in the numerator and lowest variance in the denominator with a degrees of freedom (n-1)
3. Determine the statistical hypothesis
4. State the level of significance
5. Compute the critical F value from F table. (use $\alpha/2$ for two tailed test)
6. Calculate the test statistic
7. Finally, draw the statistical conclusion. reject the null hypothesis; If the test statistic falls in the critical region.
  - p value < SL: Null hypothesis can be rejected, this means we have sufficient evidence to say that the two population variances are not equal.
  - p value > SL: Null hypothesis can not be rejected, two population variances are equal.

```
# F-Test in Python
# To perform Fishers-test we can use scipy.stats.f.cdf(f-dist-rv, numerator_dof, denomintor_dof)
import numpy as np
import pandas as pd
import scipy.stats
#define F-test function
def f_test(x, y):
    x = np.array(x)
    y = np.array(y)
    #By default, numpy.var calculates the population variance. To calculate the sample variance, we need to specify ddof=1
    f = np.var(x, ddof=1)/np.var(y, ddof=1) #calculate F test statistic , ratio of sample variance of two populations x, y
    dfn = x.size-1 #define degrees of freedom numerator 
    dfd = y.size-1 #define degrees of freedom denominator 
    p = 1-scipy.stats.f.cdf(f, dfn, dfd) #find p-value of F test statistic 
    return f, p

#perform F-test
x = [18, 19, 22, 25, 27, 28, 41, 45, 51, 55]
y = [14, 15, 15, 17, 18, 22, 25, 25, 27, 34]
f_test(x, y)
```

### Methods  of  data  collection amd Analysis of variance (ANOVA)
The way in which the data are collected determines the type of test that is used to analyze the data.
Wwe  use  an  F-ratio  that  compares  the  size  of  the  variation  due  to  our  experimental  manipulations (systematic  variation),  to  the  size  of  the  variation  due  to  random  factors (unsystematic  variation) that we cannot account for.  If  the former is big relative to the latter, we get a big value of F, and we can conclude that the observed results are unlikely to have arisen by chance. 

- **Between-group designs:** The  first  is  to  manipulate the independent variable using different participants. So, different groups of people take  part  in  each  experimental  condition  (a  between-group  or  between-subjects  design) 
  - One Way ANOVA (Independent  ANOVA)
  - $SS_{within} = SS_{error}$
  - $SS_{total} = SS_{between} + SS_{within}$

- **Repeated Measures:** The  second  method  is  to  manipulate  the  independent  variable  using  the  same  participants.  Simplistically,  this  method  means  that  we  test  the  same  group  of  participants  in  all  experimental conditions (a within-subject or repeated measures design). 
 - Repeated Measures ANOVA
 - $SS_{within} = SS_{subjects} + SS_{error}$
 - $SS_{total} = SS_{between} + SS_{within} =  SS_{between} + SS_{subjects} + SS_{error}$


## One way ANOVA

### Hypotheses

A one-way ANOVA compares the means of three or more independent (unrelated) groups to determine if there is a statistically significant difference between the corresponding population means.

\begin{align} F = \frac{\text{Between Group Variability}}{\text{Within Group Variability}}= \frac{MS_{between}} {MS_{within}} = \frac{MS_{regression}} {MS_{error}} = \frac{ns_\text{unbiased}^2} {s_\text{pooled}^2}\end{align}

**H0 (null hypothesis):** $\mu_{1} = \mu_{2} = \mu_{3} = … = \mu_{k}$ (all the population means are equal)

**H1 (alternative hypothesis):** At least one population mean is different from the rest

### Assumptions

1. Normality – Samples in each group should follow normal distribution, only needed for small sample sizes, say n < 20 per group.
2. Equal Variances – The variances of the populations that the samples come from should be equal. Homogeneity is only needed if sample sizes are very unequal (Levene's Test).
3. Independence – The observations in each group are independent of each other and the observations within groups were obtained by a random sample.
4. Same Sample Size: The treatment groups should have the same sample size.

### Naming Convention

---
- Group = Column
- i = row/replication/sample number 
- j = group/column number
- $A$ = number of observations in each groups or columns
- $B$ = number of groups or columns
- $n_{j}$ = $A$ = the sample size or number of observartions in j-th group/column
- $N = A*B$ = total number of observations
- $x_{ij}$ = observation at i-th subject (row) and j-th group(column)
- $\bar{x}_{j}$ = the mean of j-th group/column
- $GT = \sum_{i=1}^{A}\sum_{j=1}^{B} x_{ij}$ = the grand total of the entire dataset
- $\bar{x}_{grand}$ = the grand mean or overall mean of all observations
- $s_{grand}^2$ = the unbiased sample variance for the entire dataset
- $s_{i}^2$ = the unbiased sample variance for the i-th subject/row observations
- $s_{j}^2$ = the unbiased sample variance for the j-th group/column observations
- $df_{reg} = B-1$: regression degrees of freedom
- $df_{error} = N-B$: error degrees of freedom
- $df_{total} = N-1$: total degrees of freedom
- MSR = SSR/$df_{reg}$ = regression mean square
- MSE = SSE/$df_{error}$ = error mean square
- F test statistic (F = MSR/MSE)

1. **Calculate the group means and the overall mean:**

- Grand mean or overall mean
\begin{align}\bar{x}_{grand} = \frac{1}{N}\sum_{i=1}^{A}\sum_{j=1}^{B} x_{ij} = \frac{1}{N}*{GT}\end{align}

- Group mean for j-th group
\begin{align}\bar{x}_{j} = \frac{1}{n_{j}}\sum_{i=1}^{n_{j}} x_{ij} = \frac{1}{A}\sum_{i=1}^{A} x_{ij}\end{align}

2. **Total sum of squares (SST):**
- The total group variation (not variance) is comprised the sum of squares deviations of all the observations from the grand mean
- Degree of freedom is (total number of observations - 1)

Method-1
>\begin{align} SST = \sum_{i=1}^{A}\sum_{j=1}^{B}(x_{ij}-\bar{x}_{grand})^2 = \sum_{i=1}^{A}\sum_{j=1}^{B}x_{ij}^2 - \frac{{GT}^2}{N} \end{align}
 - Calculate the difference between each observation and the grand mean irrespective of group(column) or subject(row)
 - Square each of these differences
 - Add the values together to get SST

Method-2
>\begin{align} SST = (N-1)*s_{grand}^2\end{align}
\begin{align} variance_{total} = s_{grand}^2 = \frac{1}{N-1}\,\sum_{i=1}^{A}\sum_{j=1}^{B}(x_{ij}-\bar{x}_{grand})^2 \end{align}
 - Calculate the unbiased variance for all observations together irrespective of group(column) or subject(row)
 - Multiply it with (total number of observations - 1) to get SST

Method-3
>We will calculate the total sum of squares (SST) using the following formula:
\begin{align} SST = SSB + SSW = SSR + SSE \end{align}

3. **Between Groups/Samples Sum of Squares (SSB) or Regression sum of squares (SSR):** 
- Also called SS(Model) or SS(Treatment)
- Variation between Group/column means with grand means
- It accounts the variation due to the interaction between the samples/groups. If the sample means (group means) are close to each other (and therefore the grand mean) this will be small.
- The degree of freedom is equal to (number of groups or samples - 1)

Method-1
>\begin{align} SSB = \sum_{j=1}^{B} n_{j} * (\bar{x}_{j}-\bar{x}_{grand})^2 \end{align}
 - Calculate the difference between the mean of each group and the grand mean
 - Square each of these differences
 - Multiply each result by the number of participants (sample size) within that group 
 - Add the values for each group together to get SSR

Method-2
>\begin{align} SSB = \frac{1}{A}\,\sum_{j=1}^{B}(\sum_{i=1}^{A}x_{ij})^2 - \frac{GT^2}{N} \end{align}
 - Calculate the sum of observations for each group (column)
 - Square each sum (group wise sum) and add them
 - Divide it by samplesize or total number of rows, call it A
 - Sum all observations to get grand sum (grand total)
 - Square the grand total and divide it by total number of observations, call it B
 - Calculate A-B to get the SSB

4. **Within Group/Sample sum of squares (SSW) or Error sum of squares (SSE):** or SS(Residual) 
- It account for the variation due to differences within individual samples, denoted as Sum of Squares Within groups. Each sample is considered independently, no interaction between samples is involved. 
- The degrees of freedom is equal to the sum of the individual degrees of freedom for each sample. Since each sample has degrees of freedom equal to one less than their sample sizes, and there are k samples, the total degrees of freedom is k less than the total sample size: df = N - k.

Method-1
>\begin{align} SSW = \sum_{j=1}^{B}\sum_{i=1}^{A}(x_{ij}-\bar{x}_{j})^2 \end{align}
 - Calculate the difference between each observation of the group and the group mean
 - Square each of these differences
 - Add the values for each group together and call it group error SS
 - Add the group error SS of all the groups to get the SSW

Method-2
>\begin{align} SSW = \sum_{j=1}^{B}(n_{j}-1)*s_{j}^2\end{align}
\begin{align} variance_{group(j)} = s_{j}^2 = \frac{1}{A-1}\,\sum_{i=1}^{A}(x_{ij}-\bar{x}_{j})^2 \end{align}
 - Calculate the unbiased variance for j-th group observations
 - Multiply it with (total observations in the group - 1) to get SSW

5. **ANOVA table:**
Now that we have SSR, SSE, and SST, we can fill in the ANOVA table:

6. Effect Size
>\begin{align} \eta^2 = \frac{SSB}{SST} = \frac{SSR}{SST} \end{align}


7. **Interpret the results:** To determine if this is a statistically significant result, we must compare this F-statistic to the F critical value found in the F distribution table for $F(df_{reg}, df_{error})$

  - if p-value < SL,  then we reject H0 (Null Hypothesis of all means are equal)
  - if p-value >= SL,  then we fail to reject H0 (Null Hypothesis of all means are equal)


### Post-Hoc Test
In order to determine which groups are different from one another, a post-hoc test is needed

- Tukey's Honestly Significant Differences (HSD) Test
  For each pairwise comparison for sample i and j, 
  **H0 (null hypothesis):** $\mu_{i} = \mu_{j}$,
  **Ha (alternate hypothesis):** $\mu_{i} \neq \mu_{j}$

  Tukey's critical value $w = q_{\alpha}(p, df_{error}) * s_{error}$

  where 
  - $q_{\alpha}$ = Obtained from a Table of Tukey values for given significance level,
  - p = the number of groups or treatment levels
  - $n_{samplesize}$ = number of replications/observations
  - $s_{error} = \sqrt{MSE/A}$ = standard error of group mean where A is the number of observations in each group

- Rank the group means in descending order, and calculate pairwise differences. For all groups i and j, 
- if $mean_{i} - mean_{j} <= w$, it indicates the group means do not differ significantly
- if $mean_{i} - mean_{j} > w$, it indicates the group means differ significantly

```
# One way ANOVA by Statmodels Library
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
A = [12.6, 12, 11.8, 11.9, 13, 12.5, 14]
B = [10, 10.2, 10, 12, 14, 13]
C = [10.1, 13, 13.4, 12.9, 8.9, 10.7, 13.6, 12]
all_scores = A + B + C
company_names = (['A'] * len(A)) +  (['B'] * len(B)) +  (['C'] * len(C))
data = pd.DataFrame({'company': company_names, 'score': all_scores})
data.groupby('company').mean()
lm = ols('score ~ company', data = data).fit()
table = sm.stats.anova_lm(lm)
print(table)
```

```
# One way ANOVA by hand

# compute group means
group_means = data.groupby('company').mean()
#group_means = group_means.rename(columns = {'score': 'group_mean'})
group_means.columns = ['group_mean']
print("group_means",group_means)

# compute overall mean
overall_mean = data['score'].mean()
print("overall_mean=",overall_mean)

# compute Sum of Squares Total
data['overall_mean'] = overall_mean
ss_total = sum((data['score'] - data['overall_mean'])**2)
print("ss_total=",ss_total)

# add group means and overall mean to the original data frame
data = data.merge(group_means, left_on = 'company', right_index = True)

# compute Sum of Squares Error/Residual/Within
ss_residual = sum((data['score'] - data['group_mean'])**2)
print("ss_residual=",ss_residual)

# compute Sum of Squares Regression/Between/Explained/Model
ss_explained = sum((data['overall_mean'] - data['group_mean'])**2)
print("ss_explained=",ss_explained)

# compute Mean Square Error/Residual/Within
n_groups = len(set(data['company']))
n_obs = data.shape[0]
df_residual = n_obs - n_groups
ms_residual = ss_residual / df_residual
print("ms_residual=",ms_residual)

# compute Mean Square Regression/Between/Explained/Model
df_explained = n_groups - 1
ms_explained = ss_explained / df_explained
print("ms_explained=",ms_explained)

# compute F-Value
f = ms_explained / ms_residual
print("F-score=",f)

# compute p-value
import scipy.stats
p_value = 1 - scipy.stats.f.cdf(f, df_explained, df_residual)
print("p-value=",p_value)
```

## Repeated Measures ANOVA

A repeated-measures (or within-participants) test is what you use when you want to compare the performance of the same group of participants in different experimental conditions.  
That is, when the same participants take part in all of the conditions in your study.

The term one-way simply to refers to the number of independent variables (one independent variable) you have; 
- one dependent variable
- one independent variable (with 3 or more conditions)
- each participant(subject or row) takes part in all conditions/groups/columns

### Hypotheses

A repeated measures ANOVA is used to determine whether or not there is a statistically significant difference between the means of three or more groups in which the same subjects show up in each group.

1. Measuring the mean scores of subjects(group) during three or more time points.
2. Measuring the mean scores of subjects(group) under three different conditions.

### Assumptions

1. Normality: The sample values from each group are from a normally distributed population
2. Equal Variance: The population variances of the all groups should be equal (Levene's Test fro homogeneity of variance)
3. Randomness and Independence: The samples must be independent and randomly sampled
4. Same Sample Size: The treatment groups must have the same sample size.

### Naming Convention

---
- Group = Column
- Subject = Row = Block
- i = row number/subject
- j = column number/group
- $A$ = total number of subjects/rows
- $B$ = total number of groups/columns
- $n_{i}$ = $B$ = total number of observations in the i-th subject/column,
- $n_{j}$ = $A$ = total number of observations in the j-th group/column,
- $N = A*B$ = the total number of observations in the entire dataset
- $x_{ij}$ = observation at i-th subject (row) and j-th group(column)
- $\bar{x}_{i} = \frac{1}{n_{i}}\sum_{j=1}^{n_{i}} x_{ij} = \frac{1}{B}\sum_{i=1}^{B} x_{ij}$ = the mean of the i-th subject/row, 
- $\bar{x}_{j} = \frac{1}{n_{j}}\sum_{i=1}^{n_{j}} x_{ij} = \frac{1}{A}\sum_{i=1}^{A} x_{ij}$ = the mean of the j-th group/column, 
- $GT = \sum_{i=1}^{A}\sum_{j=1}^{B} x_{ij}$ = the grand total of the entire dataset
- $\bar{x}_{grand} = \frac{1}{N}\sum_{i=1}^{A}\sum_{j=1}^{B} x_{ij}$ = the grand mean or mean of the entire dataset
- $s_{grand}^2$ = the unbiased sample variance for the entire dataset
- $s_{i}^2$ = the unbiased sample variance for the i-th subject/row observations
- $s_{j}^2$ = the unbiased sample variance for the j-th group/column observations

**Example:**

In a typical one-way ANOVA, different subjects are used in each group.
But in repeated measures ANOVA, the same subjects show up in each group.

**H0 (null hypothesis):** µ1 = µ2 = µ3 (the population means are all equal)

**H1 (alternative hypothesis):** at least one population mean is different from the rest

### Steps

1. **Total sum of squares (SST):**

Method-1
>\begin{align} SST = \sum_{i=1}^{A}\sum_{j=1}^{B}(x_{ij}-\bar{x}_{grand})^2 = \sum_{i=1}^{A}\sum_{j=1}^{B}x_{ij}^2 - \frac{{GT}^2}{N} \end{align}
 - Calculate the difference between each observation and the grand mean irrespective of group(column) or subject(row)
 - Square each of these differences
 - Add the values to get SST

Method-2
>\begin{align} SST = (N-1)*s_{grand}^2\end{align}
\begin{align} variance_{total} = s_{grand}^2 = \frac{1}{N-1}\,\sum_{i=1}^{n_{subjects}}\sum_{j=1}^{B}(x_{ij}-\bar{x}_{grand})^2 \end{align}
 - Calculate the unbiased variance for all observations together irrespective of group(column) or subject(row)
 - Multiply it with (total number of observations - 1) to get SST

Method-3
> \begin{align} SST = SSB + SSW =  SSB + (SSS + SSE) \end{align} 

 - Calculate the unbiased variance for all observations together irrespective of group(column) or subject(row)
 - Multiply it with (total number of observations - 1) to get SST

2. **Between Group sum of squares or SSB(Group):** 
 - Also called SS(Group) or SS(Treatment) or SS(Model)
 - Variance of group/column mean with grand mean

Method-1
>\begin{align} SSB(Group) = \sum_{j=1}^{B}{n_{j}*(\overline{x}_{j} - \overline{x}_{grand})}^2 \end{align}
 - Calculate the difference between the mean of each group/column and the grand mean
 - Square each of these differences
 - Multiply each result by the number of observations/participants within that group
 - Add the values for each group/column together

Method-2
>\begin{align} SSB(Group) = \frac{1}{A}\,\sum_{j=1}^{B}(\sum_{i=1}^{A}x_{ij})^2 - \frac{GT^2}{N} \end{align}
 - Calculate the sum of observations for each group (column)
 - Square each sum (group wise sum) and add them
 - Divide it by total number of subjects, call it A
 - Sum all observations to get grand sum(grand total)
 - Square the grand total and divide it by total number of observations, call it B
 - Calculate A-B to get the SSB(Group)

3. **Between Subject sum of squares or SSB(Subject):** 
 - Variance of subject/row mean with grand mean

Method-1
>\begin{align} SSB(Subject) = \sum_{i=1}^{A}{n_{i}*(\overline{x}_{i} - \bar{x}_{grand})}^2 \end{align}
 - Calculate the difference between the mean of each subject/row and the grand mean
 - Square each of these differences
 - Multiply each result by the number of observations within that subject/row
 - Add the values for each group together

 Method-2
>\begin{align} SSB(Subject) = \frac{1}{B}\,\sum_{i=1}^{A}(\sum_{j=1}^{B}x_{ij})^2 - \frac{GT^2}{N} \end{align}
 - Calculate the sum of observations for each subject (row)
 - Square each sum (subject wise sum) and add them
 - Divide it by total number of groups, call it A
 - Sum all observations to get grand sum(grand total)
 - Square the grand total and divide it by total number of observations, call it B
 - Calculate A-B to get the SSB(Subject)

4. **Within Group sum of squares or SSW(Groups):** 
- Variance of group/column observations with group mean

Method-1
>\begin{align} SSW(Group) = \sum_{j=1}^{B}\sum_{i=1}^{A}(x_{ij}-\bar{x}_{j})^2 \end{align}
 - Calculate the difference between each observation of the group/column and the group/column mean
 - Square each of these differences
 - Add the values for each group together and call it group error SS
 - Add the group error SS of all the groups to get the SSW

Method-2
>\begin{align} SSW(Group) = \sum_{j=1}^{B}(n_{j}-1)*s_{j}^2\end{align}
\begin{align} variance_{group(i)} = s_{j}^2 = \frac{1}{A-1}\,\sum_{i=1}^{A}(x_{ij}-\bar{x}_{j})^2 \end{align}
 - Calculate the unbiased variance for j-th group/column observations
 - Multiply it with (total observations in the group/column - 1) to get SSW

5. **Within Subjects sum of squares or SSW(Subject):** 
- Variance of subject/row observations with subject mean

Method-1
>\begin{align} SSW(Subject) = \sum_{i=1}^{A}\sum_{j=1}^{B}(x_{ij}-\bar{x}_{i})^2 \end{align}
 - Calculate the difference between each observation of the subject/rows and the subject/row mean
 - Square each of these differences
 - Add the values for each subject/row together and call it subject error SS
 - Add the subject error SS of all the groups to get the SSW(Subject)

Method-2
>\begin{align} SSW(Subject) = \sum_{i=1}^{A}(n_{i}-1)*s_{i}^2 \end{align}
\begin{align} variance_{subject(i)} = s_{i}^2 = \frac{1}{B-1}\,\sum_{j=1}^{B}(x_{ij}-\bar{x}_{i})^2 \end{align}
 - Calculate the unbiased variance for i-th subject/row observations
 - Multiply it with (total observations in the subject/row - 1) to get SSW(Subject)

6. **Error sum of squares (SSE):** Next, we will calculate the error sum of squares (SSE), which can be found using the following formula:

\begin{align} SSE = SST – SSB(Group) – SSB(Subject) \end{align}

  - DF2 (denominator degrees of freedom) = df error = 8

We find that the F critical value is 4.459.

## Two way ANOVA

The two-away ANOVA deals with the case where there are two variables or factors and there may be interaction effects between the two factors.

You should use a two-way ANOVA when you’d like to know how two factors affect a response variable and whether or not there is an interaction effect between the two factors on the response variable.

### Hypotheses

There are three sets of hypothesis with the two-way ANOVA.
1. Main Effect - First Factor
  - **Null hypotheses(H0):** The population means of the first factor are equal. This is like the one-way ANOVA for the row factor.
  - **Alternate hypotheses(H0):** At least one population mean of the first factor is different from others.

2. Main Effect - Second Factor
  - **Null hypotheses(H0):** The population means of the second factor are equal. This is like the one-way ANOVA for the column factor.
  - **Alternate hypotheses(H0):** At least one population mean of the second factor is different from others.

3. Interaction Effect
  - **Null hypotheses(H0):** There is no interaction between the two factors. This is similar to performing a test for independence with contingency tables.
  - **Alternate hypotheses(H0):** There is some interaction between the two factors.



### Assumptions

1. Normality: The populations from which the samples were obtained - must be normally or approximately normally distributed
2. Equal Variance: The variances of the populations must be equal (Levene's Test)
3. Independence: The samples must be independent
4. Same Sample Size: The treatment groups must have the same sample size for a two-way ANOVA


### Terminology

**Factors:**
The two independent variables in a two-way ANOVA are called factors, which affect the dependent variable. Each factor will have two or more levels within it, and the degrees of freedom for each factor is one less than the number of levels.

**Treatment Groups**
Treatement Groups are formed by making all possible combinations of the two factors. SO, the total number of treatment groups is the product of the number of levels for each factor. For example, if the first factor has 3 levels and the second factor has 2 levels, then there will be 3x2=6 different treatment groups.

**Main Effect:**
The main effect involves the independent variables one at a time. The interaction is ignored for this part. Just the rows or just the columns are used, not mixed. This is the part which is similar to the one-way analysis of variance. Each of the variances calculated to analyze the main effects are like the between variances

**Interaction Effect:**
The interaction effect is the effect that one factor has on the other factor. The degrees of freedom here is the product of the two degrees of freedom for each factor.

**SS(Within) or SS(Error) Variation:**
The Within variation is the sum of squares within each treatment group. The within variance is the within variation divided by its degrees of freedom.

**F-Tests:**
There is an F-test for each of the hypotheses, and the F-test is the mean square for each main effect and the interaction effect divided by the within variance. The numerator degrees of freedom come from each effect, and the denominator degrees of freedom is the degrees of freedom for the within variance in each case.
